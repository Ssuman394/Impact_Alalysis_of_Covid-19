```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('modules are imported.')
```

```python
corona_dataset_csv = pd.read_csv("COVID-19 Dataset.csv")
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.drop(['Lat','Long'],axis = 1,inplace = True)
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

```python
corona_dataset_aggregated = corona_dataset_csv.groupby('Country/Region').sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
corona_dataset_aggregated.loc['China'].plot()
corona_dataset_aggregated.loc['India'].plot()
corona_dataset_aggregated.loc['Spain'].plot()
plt.legend()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infaction_rate = []
for country in countries:
    max_infaction_rate.append(corona_dataset_aggregated.loc[country].diff().max())
corona_dataset_aggregated['max_infaction_rate'] = max_infaction_rate
```

```python
corona_dataset_aggregated.head(10)
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated['max_infaction_rate'])
corona_data.head(10)
```

```python
world_happiness_data = pd.read_csv('Worldwide_happiness.csv')
world_happiness_data.head()
```

```python
world_happiness_data.shape
```

```python
world_happiness_data.head()
```

```python
world_happiness_data.set_index(['Country or region'],inplace = True)
world_happiness_data.head()
```

```python
data = world_happiness_data.join(corona_data).copy()
data.head()
```

```python
data.corr()
```

```python
x = data['GDP per capita']
y = data['max_infaction_rate']
sns.scatterplot(x,np.log(y))
```

```python
x = data['GDP per capita']
y = data['max_infaction_rate']
sns.regplot(x,np.log(y))
plt.title('Regression plot')
```
